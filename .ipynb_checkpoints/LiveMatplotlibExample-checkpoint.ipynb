```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook

```

```python
'''This is a lineplot (?) example'''
data = np.random.rand(100)
timeStamps = np.arange(0.0, 100, 1)

fig, ax = plt.subplots()
ax.plot(timeStamps, data)

ax.set(xlabel='time (s)', ylabel='acceleration',
       title='A read from the accelerometer vs time')
ax.grid()

# fig.savefig("test.png")
plt.show()
```
